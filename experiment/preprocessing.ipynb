```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
pd.set_option('display.max_columns', 500)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="talk")

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import pickle
from xgboost import XGBClassifier

from jcopml.plot import plot_classification_report, plot_confusion_matrix, plot_roc_curve, pl
```

## Import Dataset

```python
df = pd.read_csv("../dataset/train.csv")
```

## Train Test Split

```python
X = df.drop(columns="price_range").copy()
y = df["price_range"].copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Preprocessing X_train

```python
num_pipe = Pipeline([
    ("imputer", KNNImputer(n_neighbors=3)),
    ("scaling", MinMaxScaler())
])

cat_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessing = ColumnTransformer([
    ("numeric", num_pipe, ["battery_power", "clock_speed", "fc", "int_memory", "m_dep", "mobile_wt", "n_cores", "pc",
                                 "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time"]),
    ("categoric", cat_pipe, ["blue", "dual_sim", "four_g", "three_g", "touch_screen", "wifi"])
])

pipeline = Pipeline([
    ("prep", preprocessing),
    ("pca", PCA(n_components=16)),
    ("algo", XGBClassifier(verbosity=3, eval_metric='mlogloss'))
])
```

## Training

```python
pipeline.get_params()
```

```python
parameter_tuning = {
    'pca__n_components': np.arange(15, 19, 1),
    'algo__n_estimators': np.arange(65, 75, 1),
    'algo__gamma': np.arange(1, 3, 1),
    'algo__min_child_weight': np.arange(1, 3, 1),
    'algo__max_depth': np.arange(70, 80, 1),
    'algo__learning_rate': np.arange(0.035, 0.036, 0.00001),
    'algo__subsample': np.arange(0.8, 0.83, 0.001)
}

model = RandomizedSearchCV(pipeline, parameter_tuning, cv=4, n_jobs=-1, n_iter=100, verbose=True, random_state=42)
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))

print("\n")
print(model.best_params_)
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```

## Evaluasi

```python
plot_classification_report(X_train, y_train, X_test, y_test, model)
```

```python
plot_classification_report(X_train, y_train, X_test, y_test, model, report=True)
```

```python
plot_confusion_matrix(X_train, y_train, X_test, y_test, model)
```
